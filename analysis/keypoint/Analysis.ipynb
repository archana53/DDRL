```python
%matplotlib inline
import matplotlib.image as mpimg
import cv2 
import matplotlib.pyplot as plt 
import random
```

```python
ground_truth_file = '/home/suyash/Gatech/DDRL/modules/data/aflw_dataset/AFLW_lists/all.GTB'
with open(ground_truth_file) as file:
    ground_truths = [line.rstrip() for line in file]
```

```python
def get_random_ground_truth(ground_truths):
    selected_idx = random.randint(0, len(ground_truths))
    ground_truth = ground_truths[selected_idx]
    split_line = ground_truth.split(' ')
    return split_line[0], split_line[1]
    
```

```python
imgpath, ptspath = "/home/suyash/Gatech/DDRL/modules/data/aflw_dataset/images/3/image00168.jpg", "/home/suyash/Gatech/DDRL/modules/data/aflw_dataset/processed/annotations/3/image00168-2.pts"
```

```python
imgpath, ptspath = get_random_ground_truth(ground_truths)
print(imgpath)
print(ptspath)
```

```python
with open(ptspath) as file:
    lines = [line.rstrip() for line in file]
    
coords_x = []
coords_y = []
coords = []
for line in lines:
    split_line = line.split(' ')
    coords_x.append(split_line[1])
    coords_y.append(split_line[2])
    coords.append((split_line[1], split_line[2]))
```

```python
coords
```

```python
keypoints = [cv2.KeyPoint(float(x), float(y), size=20) for (x, y) in coords]
```

```python
  
# reading image using the imread() function 
imageread = cv2.imread(imgpath) 
  
  
# drawKeypoints function is used to draw keypoints 
output_image = cv2.drawKeypoints(imageread, keypoints, 0, (255, 0, 0), 
                                 flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS) 
  
# displaying the image with keypoints as the 
# output on the screen 
plt.imshow(output_image) 
  
# plotting image 
plt.show() 
```
